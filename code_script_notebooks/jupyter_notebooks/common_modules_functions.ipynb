```python
import random
### 1. `random.randint(a, b)`

# Generates a random integer between 1 and 10 (inclusive)
random_integer = random.randint(1, 10)
print(random_integer)

### 2. `random.choice(seq)`

# Picks a random element from the list
my_list = [1, 2, 3, 4, 5]
random_element = random.choice(my_list)
print(random_element)

### 3. `random.shuffle(lst)`

# Shuffles the elements of the list in place
my_list = [1, 2, 3, 4, 5]
random.shuffle(my_list)
print(my_list)

### 4. `random.random()`
# Generates a random float between 0 and 1
random_float = random.random()
print(random_float)

### 5. `random.uniform(a, b)`
# Generates a random float between 1 and 10
random_float = random.uniform(1, 10)
print(random_float)

### 6. `random.gauss(mu, sigma)`
# Generates a random float from a Gaussian distribution with mean 0 and standard deviation 1
random_gaussian = random.gauss(0, 1)
print(random_gaussian)

### 7. `random.sample(population, k)`

# Returns k unique elements chosen from the population
my_list = [1, 2, 3, 4, 5]
sampled_elements = random.sample(my_list, 3)
print(sampled_elements)

### 8. `random.randrange(start, stop, step)`

# Generates a random integer from the range 0 to 10 (exclusive) with a step of 2
random_integer = random.randrange(0, 10, 2)
print(random_integer)

### 9. `random.seed(a=None, version=2)`

# Sets the seed for the random number generator
random.seed(42)

# Now, the random operations will generate the same sequence each time
print(random.random())
print(random.randint(1, 10))

### 10. `random.getstate()`, `random.setstate(state)`

# Get the current internal state of the random number generator
state = random.getstate()

# Set the state back to reproduce the same sequence of random numbers
random.setstate(state)
print(random.random())
```

```python
# 5. **`sys` Module:** - `sys.argv` - `sys.exit([arg])` - `sys.stdin` - `sys.stdout` - `sys.stderr` - `sys.platform` - `sys.version` - `sys.path` - `sys.modules` - `sys.getsizeof(object[, default])`

import sys

def main():
    # 1. sys.argv: Command-line arguments passed to a script.
    print("Command-line arguments:", sys.argv)

    # 2. sys.exit([arg]): Exits from Python, optional arg can be returned to the calling environment.
    if len(sys.argv) < 2:
        print("Please provide a command-line argument.")
        sys.exit(1)  # Exit with a non-zero status to indicate an error

    # 3. sys.stdin: Standard input
    print("Enter something:")
    user_input = sys.stdin.readline().strip()
    print("You entered:", user_input)

    # 4. sys.stdout: Standard output
    print("This is a message on standard output.")

    # 5. sys.stderr: Standard error
    sys.stderr.write("This is an error message.\n")

if __name__ == "__main__":
    main()
```

```python
# 6. **`json` Module:** - `json.dumps(obj, indent=None)` - `json.loads(s)` - `json.dump(obj, fp)` - `json.load(fp)` - `json.JSONEncoder` - `json.JSONDecoder` - `json.JSONEncoder.default(o)` - `json.JSONDecoder.object_hook(dct)` - `json.JSONEncoder.encode(o)` - `json.JSONDecoder.decode(s)`

import json

# Example data
data = {
    "name": "John Doe",
    "age": 30,
    "city": "New York",
    "is_student": False,
    "grades": [90, 85, 92]
}

# 1. json.dumps(obj, indent=None)
json_string = json.dumps(data, indent=2)
print("JSON String:")
print(json_string)
print()

# 2. json.loads(s)
json_data = json.loads(json_string)
print("Decoded JSON Data:")
print(json_data)
print()

# 3. json.dump(obj, fp)
with open("example.json", "w") as json_file:
    json.dump(data, json_file, indent=2)
print("Data written to 'example.json'")
print()

# 4. json.load(fp)
with open("example.json", "r") as json_file:
    loaded_data = json.load(json_file)
print("Loaded Data from 'example.json':")
print(loaded_data)
print()

# Custom JSONEncoder and JSONDecoder
class MyEncoder(json.JSONEncoder):
    def default(self, o):
        if isinstance(o, set):
            return list(o)
        return super().default(o)

class MyDecoder(json.JSONDecoder):
    def object_hook(self, dct):
        if "_set" in dct:
            return set(dct["_set"])
        return dct

# 5. json.JSONEncoder
encoded_custom_data = MyEncoder().encode({"colors": {"red", "green", "blue"}})
print("Custom Encoded JSON Data:")
print(encoded_custom_data)
print()

# 6. json.JSONDecoder
decoded_custom_data = MyDecoder().decode('{"colors": {"_set": ["red", "green", "blue"]}}')
print("Custom Decoded JSON Data:")
print(decoded_custom_data)
```

```python
# 7. **`requests` Module:** - `requests.get(url, params=None, **kwargs)` - `requests.post(url, data=None, json=None, **kwargs)` - `requests.put(url, data=None, **kwargs)` - `requests.delete(url, **kwargs)` - `requests.request(method, url, **kwargs)` - `requests.Response` - `requests.Response.status_code` - `requests.Response.headers` - `requests.Response.json()` - `requests.Response.text`
import requests

# Example URLs
get_url = "https://jsonplaceholder.typicode.com/posts/1"
post_url = "https://jsonplaceholder.typicode.com/posts"
put_url = "https://jsonplaceholder.typicode.com/posts/1"
delete_url = "https://jsonplaceholder.typicode.com/posts/1"

# 1. GET Request
response_get = requests.get(get_url)
print(f"GET Response Status Code: {response_get.status_code}")
print("GET Response Headers:")
print(response_get.headers)
print("GET Response JSON:")
print(response_get.json())
print("GET Response Text:")
print(response_get.text)
print("\n" + "-"*50 + "\n")

# 2. POST Request
post_data = {
    "title": "foo",
    "body": "bar",
    "userId": 1
}
response_post = requests.post(post_url, json=post_data)
print(f"POST Response Status Code: {response_post.status_code}")
print("POST Response Headers:")
print(response_post.headers)
print("POST Response JSON:")
print(response_post.json())
print("POST Response Text:")
print(response_post.text)
print("\n" + "-"*50 + "\n")

# 3. PUT Request
put_data = {
    "title": "foo",
    "body": "bar",
    "userId": 1
}
response_put = requests.put(put_url, json=put_data)
print(f"PUT Response Status Code: {response_put.status_code}")
print("PUT Response Headers:")
print(response_put.headers)
print("PUT Response JSON:")
print(response_put.json())
print("PUT Response Text:")
print(response_put.text)
print("\n" + "-"*50 + "\n")

# 4. DELETE Request
response_delete = requests.delete(delete_url)
print(f"DELETE Response Status Code: {response_delete.status_code}")
print("DELETE Response Headers:")
print(response_delete.headers)
print("DELETE Response JSON:")
# Since DELETE might not return JSON, we print an informative message
print("DELETE Response Text:")
print(response_delete.text if response_delete.text else "No content returned.")
```

```python
# 8. **`re` Module:** - `re.match(pattern, string, flags=0)` - `re.search(pattern, string, flags=0)` - `re.findall(pattern, string, flags=0)` - `re.finditer(pattern, string, flags=0)` - `re.sub(pattern, repl, string, count=0, flags=0)` - `re.compile(pattern, flags=0)` - `re.split(pattern, string, maxsplit=0, flags=0)` - `re.escape(string)` - `re.fullmatch(pattern, string, flags=0)` - `re.IGNORECASE`, `re.MULTILINE`, `re.DOTALL`, etc.
import re

def demonstrate_re_functions():
    # Example string
    input_string = "Hello, World! This is an example string for regex testing. world is different from World."

    # Function 1: re.match(pattern, string, flags=0)
    match_result = re.match(r"Hello", input_string)
    print("Match Result:", match_result.group() if match_result else "No match")

    # Function 2: re.search(pattern, string, flags=0)
    search_result = re.search(r"World", input_string)
    print("Search Result:", search_result.group() if search_result else "Not found")

    # Function 3: re.findall(pattern, string, flags=0)
    findall_result = re.findall(r"World", input_string)
    print("Findall Result:", findall_result)

    # Function 4: re.finditer(pattern, string, flags=0)
    finditer_result = [match.group() for match in re.finditer(r"World", input_string)]
    print("Finditer Result:", finditer_result)

    # Function 5: re.sub(pattern, repl, string, count=0, flags=0)
    substituted_string = re.sub(r"World", "Universe", input_string, flags=re.IGNORECASE)
    print("Substituted String:", substituted_string)

    # Function 6: re.compile(pattern, flags=0)
    regex_pattern = re.compile(r"example", flags=re.IGNORECASE)
    compiled_result = regex_pattern.search(input_string)
    print("Compiled Result:", compiled_result.group() if compiled_result else "Not found")

    # Function 7: re.split(pattern, string, maxsplit=0, flags=0)
    split_result = re.split(r"\W+", input_string)
    print("Split Result:", split_result)

    # Function 8: re.escape(string)
    special_chars = re.escape(".*^$")
    print("Escaped Special Characters:", special_chars)

    # Function 9: re.fullmatch(pattern, string, flags=0)
    fullmatch_result = re.fullmatch(r"Hello, World!.*Universe.*string", substituted_string)
    print("Fullmatch Result:", "Match" if fullmatch_result else "No match")

    # Function 10: Flags - re.IGNORECASE, re.MULTILINE, re.DOTALL
    flags_result = re.search(r"world.*example", input_string, flags=re.IGNORECASE | re.DOTALL)
    print("Flags Result:", flags_result.group() if flags_result else "Not found")

demonstrate_re_functions()
```

```python
# 9. **`collections` Module:** - `collections.Counter(iterable)` - `collections.defaultdict(default_factory)` - `collections.namedtuple(typename, field_names, *, rename=False, defaults=None, module=None)` - `collections.OrderedDict([items])` - `collections.deque([iterable[, maxlen]])` - `collections.ChainMap(*maps)` - `collections.UserDict([dict])` - `collections.UserList([list])` - `collections.UserString([sequence])` - `collections.abc`

from collections import Counter, defaultdict, namedtuple, OrderedDict, deque, ChainMap, UserDict, UserList, UserString

# Counter
numbers = [1, 2, 3, 4, 1, 2, 1, 5, 3, 4, 6, 7, 7, 7]
counter = Counter(numbers)
print("Counter:", counter)

# defaultdict
fruit_count = defaultdict(int)
fruits = ['apple', 'orange', 'banana', 'apple', 'banana', 'pear', 'orange', 'apple']
for fruit in fruits:
    fruit_count[fruit] += 1
print("DefaultDict:", fruit_count)

# namedtuple
Person = namedtuple('Person', ['name', 'age', 'gender'])
person = Person(name='Alice', age=25, gender='Female')
print("NamedTuple:", person)

# OrderedDict
ordered_dict = OrderedDict()
ordered_dict['one'] = 1
ordered_dict['three'] = 3
ordered_dict['two'] = 2
print("OrderedDict:", ordered_dict)

# deque
d = deque([1, 2, 3, 4, 5], maxlen=3)
print("Deque:", d)

# ChainMap
dict1 = {'one': 1, 'two': 2}
dict2 = {'three': 3, 'four': 4}
chain_map = ChainMap(dict1, dict2)
print("ChainMap:", chain_map)

# UserDict
user_dict = UserDict({'a': 1, 'b': 2})
print("UserDict:", user_dict)

# UserList
user_list = UserList([1, 2, 3, 4, 5])
print("UserList:", user_list)

# UserString
user_string = UserString("Hello, World!")
print("UserString:", user_string)

# collections.abc
from collections.abc import MutableSequence
print("Is MutableSequence:", isinstance(user_list, MutableSequence))
```

```python
# 10. **`csv` Module:** - `csv.reader(csvfile, dialect='excel', **fmtparams)` - `csv.writer(csvfile, dialect='excel', **fmtparams)` - `csv.DictReader(csvfile, fieldnames=None, restkey=None, restval=None, **fmtparams)` - `csv.DictWriter(csvfile, fieldnames, restval='', extrasaction='raise', **fmtparams)` - `csv.field_size_limit([new_limit])` - `csv.get_dialect(name)` - `csv.list_dialects()` - `csv.unregister_dialect(name)` - `csv.register_dialect(name[, dialect[, **fmtparams]])` - `csv.QUOTE_ALL`, `csv.QUOTE_MINIMAL`, `csv.QUOTE_NONNUMERIC`, `csv.QUOTE_NONE`

import csv

# Example data for writing to a CSV file
data_to_write = [
    ['Name', 'Age', 'City'],
    ['Alice', 25, 'New York'],
    ['Bob', 30, 'San Francisco'],
    ['Charlie', 22, 'Los Angeles']
]

# Example 1: Writing to a CSV file using csv.writer
with open('example.csv', 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for row in data_to_write:
        csv_writer.writerow(row)

# Example 2: Reading from a CSV file using csv.reader
with open('example.csv', 'r', newline='') as csvfile:
    csv_reader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in csv_reader:
        print(row)

# Example 3: Writing to a CSV file using csv.DictWriter
data_to_write_dict = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'San Francisco'},
    {'Name': 'Charlie', 'Age': 22, 'City': 'Los Angeles'}
]

fieldnames = ['Name', 'Age', 'City']

with open('example_dict.csv', 'w', newline='') as csvfile:
    csv_dict_writer = csv.DictWriter(csvfile, fieldnames=fieldnames, extrasaction='raise', delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    csv_dict_writer.writeheader()
    csv_dict_writer.writerows(data_to_write_dict)

# Example 4: Reading from a CSV file using csv.DictReader
with open('example_dict.csv', 'r', newline='') as csvfile:
    csv_dict_reader = csv.DictReader(csvfile, delimiter=',', quotechar='"')
    for row in csv_dict_reader:
        print(row)

# Example 5: Other csv module functions
print(f"CSV Field Size Limit: {csv.field_size_limit()}")
print(f"Available Dialects: {csv.list_dialects()}")
print(f"Current Dialect: {csv.get_dialect('excel')}")

# Note: Registering a new dialect is typically not required for common use cases.
# You can register a new dialect if you have specific formatting requirements.
# csv.register_dialect('custom', delimiter='|', quoting=csv.QUOTE_ALL)
# Then use it like: csv.writer(file, dialect='custom')
```
