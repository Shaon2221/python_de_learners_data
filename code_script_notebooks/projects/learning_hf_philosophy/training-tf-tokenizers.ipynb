Lets build a tokenizer from scratch. Well that is gonna be very involved

The pipeline involves the following, 

- Normalisation: series of steps like uncasing, removing spaces, unicode normalisation 

- Pre-Tokenization: Which decides how to pre-segment the sentences

- Model: Trainable part, which handles the token discovery and generation

- Post-Processing: Advanced construction features, that is required by SoTA models

- Decoding : Mapping original token back to string

```python
# Building WordPiece model like Bert

from tokenizers import (
    decoders,
    models,
    normalizers,
    pre_tokenizers,
    processors,
    trainers,
    Tokenizer
)
```

```python
# instantiating wordpiece tokenizer 

tokenizer = Tokenizer(models.WordPiece(unk_token="[UNK]"))
bpe_tokenizer = Tokenizer(models.BPE()) # Byte Pair Encoding Tokenizer
uni_tokenizer = Tokenizer(models.Unigram()) # Unigram 

# adding normalizers and pre-tokenizers

tokenizer.normalizer = normalizers.BertNormalizer(lowercase=True)
# No normalizer for BPE
uni_tokenizer.normalizer = normalizers.Sequence(
[normalizers.Replace("`",'"'), normalizers.Replace("''", '"'),
normalizers.Lowercase()]
)

# tokenizer.normalizer = normalizers.Sequence(
#    [normalizers.NFD(), normalizers.Lowercase(), normalizers.StripAccents()])

tokenizer.pre_tokenizer = pre_tokenizers.BertPreTokenizer()
bpe_tokenizer.pre_tokenizer = pre_tokenizers.ByteLevel(add_prefix_space=False)
uni_tokenizer.pre_tokenizer = pre_tokenizers.Metaspace()  # replace spaces with _
```

```python
tokenizer.pre_tokenizer.pre_tokenize_str("This is a test string")
```

```python
bpe_tokenizer.pre_tokenizer.pre_tokenize_str('THis is a test')
```

```python
uni_tokenizer.pre_tokenizer.pre_tokenize_str("This is a test")
```

```python
# attaching the trainer

special_tokens = ["[UNK]","[PAD]","[CLS]","[SEP]","[MASK]"]
trainer = trainers.WordPieceTrainer(vocab_size=25000,
                                   special_tokens=special_tokens)
```

```python
bpe_trainer = trainers.BpeTrainer(vocab_size=25000,
                                 special_tokens=["<|endoftext|>"])
bpe_tokenizer.train_from_iterator(iter(wikiloader), trainer=bpe_trainer) 

# next the post_processor will be created, after wordpiece tokenizer post processing
```

```python
uni_trainer = trainers.UnigramTrainer(vocab_size=25000,
                                     special_tokens=["[CLS]","[SEP]",
                                                    "<unk>", "<pad>","[MASK]"])
uni_tokenizer.train_from_iterator(iter(wikiloader),
                                  trainer=uni_trainer)
```

```python
# train the tokenizer
tokenizer.train_from_iterator(iter(wikiloader), trainer=trainer)
```

```python
# define the post processing for word-piece
cls_token_id = tokenizer.token_to_id("[CLS]")
sep_token_id = tokenizer.token_to_id("[SEP]")
print(cls_token_id, sep_token_id)
```

```python
# define the post processing for uni-gram
cls_token_id = uni_tokenizer.token_to_id("[CLS]")
sep_token_id = uni_tokenizer.token_to_id("[SEP]")
unk_token_id = uni_tokenizer.token_to_id("<unk>")
print(cls_token_id, sep_token_id, unk_token_id)
```

```python
tokenizer.encode("Lets test it out").tokens
```

```python
tokenizer.encode("Lets move this to ides").ids
```

```python
# here is how we can build our post processor for word-piece. 
# We have to indicate in the template 
# how to organize the special tokens with one sentence 
# ($A) or two sentences ($A and $B). The : followed by a number indicates the token 
# type ID to give to each part.

tokenizer.post_processor = processors.TemplateProcessing(
    single=f"[CLS]:0 $A:0 [SEP]:0",
    pair=f"[CLS]:0 $A:0 [SEP]:0 $B:1 [SEP]:1",
    special_tokens=[
        ("[CLS]", cls_token_id),
        ("[SEP]", sep_token_id)
    ],
)

tokenizer.decoder = decoders.WordPiece(prefix="##")
```

```python
uni_tokenizer.post_processor = processors.TemplateProcessing(
    single=f"[CLS]:0 $A:0 [SEP]:0",
    pair=f"[CLS]:0 $A:0 [SEP]:0 $B:1 [SEP]:1",
    special_tokens=[
        ("[CLS]", cls_token_id),
        ("[SEP]", sep_token_id)
    ],
)

uni_tokenizer.decoder = decoders.Metaspace()
```

```python
# BPE post processing
bpe_tokenizer.post_processor = processors.ByteLevel(trim_offsets=False)
bpe_tokenizer.decoder = decoders.ByteLevel()
```
